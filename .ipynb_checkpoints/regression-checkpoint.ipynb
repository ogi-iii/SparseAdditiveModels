```python
import numpy as np
import pandas as pd
import scipy
from scipy.interpolate import UnivariateSpline
from sklearn.neighbors import KNeighborsRegressor
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pylab as plt
%matplotlib inline
```

# data loading

```python
# sample data
boston = load_boston()
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
boston_df['MED_PRICE'] = boston.target

boston_df.head()
```

```python
X = boston_df.drop("MED_PRICE", axis=1).values
y = boston_df["MED_PRICE"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# fitting to SpAM

```python
# param
lambda_ = 2
max_iter = 1000
threshold = 5
```

```python
def positive_part(val):
    if val > 0:
        pp = val
    else:
        pp = 0
        
    return pp
```

```python
def setting(X, y):
    
    if X.shape[0] == y.shape[0]:
        f = np.zeros_like(X)
        
        for j in range(X.shape[1]):
            x_idx = np.arange(X.shape[0])
            us = UnivariateSpline(x_idx, X[:, j])
            f[:, j] = us(x_idx) - np.mean(X[:, j])
            
        ave_y = np.mean(y)
        return f, ave_y
        
    else:
        print("DATA SHAPE ERROR!: Data shape must be X.shape[0] == y.shape[0]")
```

```python
def fit(X, y):
    f, ave_y = setting(X_train, y_train) # init setting
    
#     pre_mse = mean_squared_error(y, ave_y - np.sum(f, axis=1))

    for _ in range(max_iter):

        for j in np.random.permutation(np.arange(f.shape[1])):

            f_k = np.delete(np.copy(f), j, axis=1)
            R_j = y - ave_y - np.sum(f_k, axis=1) # compute the residual

            S_j = UnivariateSpline(np.arange(len(R_j)), R_j)
            P_j = S_j(np.arange(len(R_j))) # estimate the projection by spline smoothing

            s_j = np.sqrt(np.mean(P_j**2)) # estimate the norm by (15)

            f_j = positive_part(1 - (lambda_ / s_j)) * P_j # soft-threshold

            f_j = f_j - np.mean(f_j) # center

            f[:, j] = f_j
            
#         if np.abs(pre_mse - mean_squared_error(y, ave_y - np.sum(f, axis=1))) <= threshold:
#             break
#         else:
#             pre_mse = mean_squared_error(y, ave_y - np.sum(f, axis=1))
    
    return f, ave_y
```

```python
%%time
f, ave_y = fit(X_train, y_train)
```

```python
pd.DataFrame(f, columns=boston.feature_names).head()
```

```python
for i in range(f.shape[1]):
    plt.title(f'{boston_df.columns[i]}')
    plt.plot(f[:, i])
    plt.show()
    
pred = ave_y + np.sum(f, axis=1)
plt.plot(y_train, "ro")
plt.plot(pred)
```

# prediction from SpAM

```python
def prediction(X_train, y_train, X_test, f, ave_y):
    
    fxs = np.zeros_like(X_test)
    
    for i in range(X_train.shape[1]):
        
        knn = KNeighborsRegressor().fit(X_train[:, i].reshape(-1,1), f[:, i]) # connect f with X_train  
        fxs[:, i] = knn.predict(X_test[:, i].reshape(-1,1)) # calculate fxs from X_test
        
    pred = ave_y + np.sum(fxs, axis=1)
    
    return pred
```

```python
pred = prediction(X_train, y_train, X_test, f, ave_y)
```

```python
plt.plot(y_test, "ro")
plt.plot(pred)
```

```python
r2_score(y_test, pred)
```
